```python
def add(*x):
    sum_x = 0
    for i in x:
        sum_x += i
    print(f'The cummulative sum of the numbers is {sum_x}.')

def subtract(*x):
    sum_x = 0
    for i in x:
        sum_x -= i
    print(f'The cummulative difference of the numbers is {sum_x}.')

def multiply(*x):
    sum_x = 1
    for i in x:
        sum_x *= i
    print(f'The cummulative product of the numbers is {sum_x}.')

def divide(*x):
    sum_x = 1
    for i in x:
        sum_x /= i
    print(f'The cummulative quotient of the numbers is {sum_x}.')

def cacn(x):
    return any(i.isalpha() for i in x) and any(i.isnumeric() for i in x)

def sign_up():
    fname = input('Enter your First name>>> ')
    lname = input('Enter your last name>>> ')
    ename = input('Enter your email>>> ')
    usname = input('Enter a username for login purposes>>> ')
    pword1 = input('Enter a password that is alphanumeric>>> ')
    if cacn(pword1) == True:
        pword2 = input('Please enter the password again>>> ')
        if pword1 == pword2:
            gotten = [fname,lname,ename,usname,pword2]
            with open(gotten[0] + '_nami.txt', 'w') as file:
                for i in gotten:
                    file.write(i + '\n')
            print('You are good to go')
            #return gotten
        else:
            print('Password does not match')
    else:
        print('Invalid Password! Mix numbers and alphabets.')
        
def login(x):
    usname = input('Enter your username>>> ')
    pword = input('Enter your password>>> ')
    if usname == x[3] and pword == x[4]:
        print('You are welcome, ',x[0], x[1])
        return True
    elif usname != x[3] and pword == x[4]:
        print('Your username is incorrect.')
    elif usname == x[3] and pword != x[4]:
        print('Your password is incorrect.')
    else:
        print('We do not have your record.')
	
#x = sign_up()


def get_file():
	nam = input('What is your name?>>> ')
	with open(nam + '_nami.txt', 'rt') as file:
		net = file.read().strip('\n').split()
		print(net)
		return net

def menu():
    print('WELCOME TO THIS LOGIN IMPLEMENTATION'.center(50, "*"))
    print('Please make a choice:\n\t1. Instructions\n\t2. Sign up\n\t3. Login\n\t4. Quit ')
    choice = input('Enter your choice>>> ')
    return choice
#get_file()
#x = get_file()
#login(x)

# Running the program
res = menu()
while res != '5':
    if res == "1":
        print("These are the instructions")
    elif res == "2":
        sign_up()
    elif res == "3":
        try:
            x = get_file()
            y = login(x)
            if y:
                step = "0"
                print("1. Addition\n2. Subtraction\n3. Multiplication\n4. Division\n5. Exit")
                step = input("Choose the task: ")
                while step != "0":
                    if step == "1":
                        num = input('Enter the numbers to be added separated by coma>>> ')
                        num = num.split(',')
                        num = [int(i) for i in num]
                        add(*num)
                    elif step == "2":
                        num = input('Enter the numbers to be subtracted separated by coma>>> ')
                        num = num.split(',')
                        num = [int(i) for i in num]
                        subtract(*num)
                    elif step == "3":
                        num = input('Enter the numbers to be multiplied separated by coma>>> ')
                        num = num.split(',')
                        num = [int(i) for i in num]
                        multiply(*num)
                    elif step == "4":
                        num = input('Enter the numbers to be divided separated by coma>>> ')
                        num = num.split(',')
                        num = [int(i) for i in num]
                        divide(*num)
                    elif step == "5":
                        print("Exiting Menu")
                        break
                    step = input("Choose the task: ")
        except:
            print("Your name doesn't sound familiar. Please check that it is correctly typed.")
            
    elif res == "4":
        print("Goodbye!")
        break
    res = menu()
    
```
